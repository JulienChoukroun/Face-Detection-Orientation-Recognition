Transform a video in frames.

```
import cv2
 
# Open the video file
cap = cv2.VideoCapture('/content/sq76096_2.avi')
i = 0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    cv2.imwrite('/content/Images/image'+str(i)+'.jpg',frame)
    i+=1

# maximum is the number of frames
maximum = i
print(maximum)

cap.release()
cv2.destroyAllWindows()
```

Test for all the frames with the execution time.

We detect only the points of faces.

```
# running offline
%%bash
start=`date +%s`
var=`cd /content/Images && ls -l | wc -l`;
i=0;
while [ $i -lt $var ]; do
  a="/content/Images/image";
  b=$i;
  c=".jpg";
  # Use 2d_sparse to detect the points of faces
  for OPT in 2d_sparse; do
    python demo.py -f "${a}${b}${c}" -o $OPT --show_flag=false --onnx;
  done;
  ((i++));
done;
end=`date +%s`
runtime=$((end-start))
echo $start
echo $end
echo $runtime
```

Test for all the frames with the execution time.

We detect the orientation of the faces. If the person is looking at the camera, the parallelepiped will be colored green. Otherwise, the figure will be colored red.

```
# running offline
%%bash
start=`date +%s`
var=`cd /content/Images && ls -l | wc -l`;
i=0;
while [ $i -lt $var ]; do
  a="/content/Images/image";
  b=$i;
  c=".jpg";
  # Use pose to detect the orientation of the faces
  for OPT in pose; do
    python demo.py -f "${a}${b}${c}" -o $OPT --show_flag=false --onnx;
  done;
  ((i++));
done;
end=`date +%s`
runtime=$((end-start))
echo $start
echo $end
echo $runtime
```

Transform the set of frames in a video.

```
import cv2
import numpy as np

img_array = []
i = 1000
while(i < 2200):
    filename= '/content/ImagesAlgo/image'+str(i)+'_pose.jpg'
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)
    i+=1

out = cv2.VideoWriter('/content/VideoTramReconstruite13.mp4',cv2.VideoWriter_fourcc(*'MP4V'), 15.107566, size)
 
for i in range(len(img_array)):
    out.write(img_array[i])
    cv2.waitKey(50)
out.release()
```
